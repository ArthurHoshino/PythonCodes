```python
# Automação de Sistemas e Processos com Python

# Passo a passo
# Passo 1: Entrar no sistema da empresa (no caso da aula, será a pasta do drive com o arquivo)
# Passo 2: Navegar no sistema e encontrar a base de dados (entrar na pasta "Exportar")
# Passo 3: Download da base de dados
# Passo 4: Calcular os indicadores (faturamento, quantidade de produtos)
# Passo 5: Entrar no email
# Passo 6: Mandar um email para a diretoria com os indicadores

import webbrowser
import pyautogui
import pyperclip
import time
import pandas as pd

pyautogui.PAUSE = 0.8
url = "https://google.com"

#abrir uma nova aba e acessar a base de dados
webbrowser.open(url, new=0)
time.sleep(3)
# pyautogui.hotkey("ctrl", "t")
pyautogui.hotkey("ctrl", "e")
pyperclip.copy("https://drive.google.com/drive/folders/149xknr9JvrlEnhNWO49zPcw0PW5icxga")
pyautogui.hotkey("ctrl", "v")
pyautogui.press("enter")
time.sleep(3)

# abrir a pasta da base de dados
pyautogui.click(x=430, y=283, clicks=2)
time.sleep(2)

# fazer o download da base de dados
pyautogui.click(x=448, y=376, button='right')
pyautogui.click(x=566, y=861)
time.sleep(7)

# salvar a informação do arquivo Excel na variável usando o pandas
tabela = pd.read_excel(r"C:\Users\Arthur\Downloads\Vendas - Dez.xlsx") # "r" serve para ignorar a formatação de "\"

# soma da coluna "quantidade" e "valor final"
quantidade = tabela['Quantidade'].sum()
faturamento = tabela['Valor Final'].sum()

# abir o email
pyautogui.hotkey("ctrl", "t")
pyautogui.write("https://gmail.com")
pyautogui.press("enter")

pyautogui.click(x=16, y=157) # esse só para o meu caso, para fechar a aba de download
time.sleep(5)

# botão de escrever novo email
pyautogui.click(x=119, y=170)

# destinatário
pyautogui.write("natalliestragnolli@gmail.com")
pyautogui.press("tab")
pyautogui.press("tab")

# assunto
pyperclip.copy("Relatório de Vendas")
pyautogui.hotkey("ctrl", "v")
pyautogui.press("tab")

# corpo do email
texto = f"""Prezada Natallie,
O faturamento do mês foi de: R$ {faturamento}
A quantidade de produtos foi de: {quantidade}

É isso,
Arthur""" # o "f" é para formatacao e ele aceitar variaveis dentro do texto

pyperclip.copy(texto)
pyautogui.hotkey("ctrl", "v")
time.sleep(2)

# enviar o email
pyautogui.hotkey("ctrl", "enter")
```
